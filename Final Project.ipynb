Importing all packs.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

Creating an array to store the data.

```python
totalArray = np.ndarray(shape=(38,6))
#print(totalArray)
```

Defining the function which will take the average temperature and store it in the totalArray.

```python
def tempData(foldName, colNum):
    rowIndex = 0
    for filename in os.listdir(foldName):
        if filename.endswith(".csv"):
            fname = filename
            data = np.genfromtxt(foldName+fname, delimiter = ',', skip_header = 1)
            totalArray[rowIndex,colNum] = np.average(data[:,13])
            rowIndex = rowIndex + 1
    return;
```

Creating a 3D plot of all the resulting data.

```python
x = (20,100,250,500,1000,1500)
y = range(38)

data = np.transpose(totalArray)

plotArea = plt.figure(figsize=(20.,10.))
Plot = plotArea.add_subplot(111, projection='3d')

X, Y = np.meshgrid(x, y)  # `plot_surface` expects `x` and `y` data to be 2D
Plot.plot_surface(X.T, Y.T, data)

Plot.set_xlabel('Depth in Meters')
Plot.set_ylabel('Time in Months')
Plot.set_zlabel('Temperature in Degrees C')

plt.show()
```

Defining a new function to plot the temperature profile at specific depths as time changes.

```python
def plotProfT(colNum,depth):
    x = range(0,38)
    y = totalArray[:,colNum]
    plt.plot(x,y,label=depth)
    plt.xlabel('Time')
    plt.ylabel('Average Temperature')
    plt.legend(loc='upper left')
    #plt.ylim(0,20)
```

```python
fig=plt.figure(figsize=(15, 10))
plt.title('Temperature Profile with Time at Certain Depths')
plotProfT(0,20)
plotProfT(1,100)
plotProfT(2,250)
plotProfT(3,500)
plotProfT(4,1000)
plotProfT(5,1500)
```
